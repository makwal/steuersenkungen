# Code ESTV_3 - 4. Teilauswertung: Steuerbelastung juristische Personen 2020

In diesem Notebook wird der Frage nachgegangen, wie die durchschnittliche Steuerbelastung für juristische Personen in den Kantonen aussieht. Die Hauptdaten des Projekts gehen ja nur bis 2016, was nicht gerade aktuell ist. Deshalb werden die 2020er-Daten der Kantone noch beigezogen. Es wird explizit nur das Jahr 2020 beleuchtet und nicht die Entwicklung 2016-2020. Dies aus zwei Gründen: Zum einen liessen sich diese Daten nicht mit jenen aus dem anderen Bericht vergleichen, der den Zeitraum 2004 bis 2016, weil die Datenbasis nicht genau die gleiche ist (zB. hier ist auch die Bundessteuer dabei). Zum anderen gibt es für die natürlichen Personen keine vergleichbaren Daten für den Zeitraum 2016 bis 2020. Kantonsvergleiche sind aber stets wichtig, wenn es um Steuerbelastungen geht, da die Kantone in einem Wettbewerb um das Steuersubstrat der Firmen stehen. Deshalb der Fokus auf 2020.

Für diesen Vergleich werden Daten aus einem Bericht der Eidgenössischen Steuerverwaltung aus dem Jahr 2020 beigezogen (zu finden im Ordner Berichte_estv). Für ein gemäss Bericht repräsentatives Unternehmen (260'000.- Gewinn, 2 Mio Eigenkapital) wird die durchschnittliche Steuerbelastung pro Kanton errechnet. Die Daten liegen auf Ebene der Gemeinden vor. Die Steuerbelastungen werden mit dem Umfang des Steuersubstrats pro Gemeinde gewichtet (Variable nettosteuer_fr), damit jede gemäss ihrer wirtschaftlichen Bedeutung in die Auswertung einfliesst.

Erklärung der Spalten des Dataframes mit den Belastungsdaten:

- Jahr
- kanname (Kantonskürzel)
- gdenr (Bfs-Nummer)
- gname_bfs (Gemeindename)
- k (Kapital)
- y (Reingewinn)
- etr (Effective tax rate; effektive Steuerbelastung in Prozent)

```python
import pandas as pd
%matplotlib inline
```

```python
#df für die Belastungsdaten, df_g für die Gewichtungsdaten
df = pd.read_csv('Daten/estv_juristische_2003-2020.csv')
df_g = pd.read_excel('Daten/estv_steuersubstrat_gemeinden.xlsx', sheet_name='Daten')
```

```python
#Nur die notwendigen Spalten mitnehmen, sprich nur Einkommen (y) 260000 und Kapital (k) 2 Millionen
del df['kannr']
df = df[(df['k'] == 2.000000e+06) & (df['y'] == 260000)].copy()
df_g = df_g[['jahr', 'gem_nr', 'gem_bez', 'nettosteuer_fr']].copy()
```

Im übernächsten Schritt werden die beiden Dataframes gemergt. Dies geschieht anhand der Datumsspalte ('Jahr') und der Bfs-Nummer. Im df_g sind die Jahresangaben aber verkürzt (3 = 2003, 17 = 2017). Deshalb wird über die Funktion 'jahr_fixer' jede Jahreszahl korrigiert.

```python
def jahr_fixer(elem):
    return elem + 2000

df_g['jahr'] = df_g['jahr'].apply(jahr_fixer)
```

```python
df_g.head()
```

```python
#Check, ob die Jahresangaben stimmen
df_g['jahr'].unique()
```

```python
#Merge anhand der Spalten Bfs-Nummer (oder Gemeinde-Nr) und Jahr
df = df.merge(df_g, left_on=['gdenr', 'jahr'], right_on=['gem_nr', 'jahr'], how='left')

#Spalten sortieren nach Bfs-Nummer und Jahr
df.sort_values(by=['gdenr', 'jahr'], inplace=True)
```

```python
#Nicht benötigte Spalten löschen
del df['gem_nr']
del df['gem_bez']
```

```python
df.head(30)
```

Die Gewichtungsdaten reichen nur bis ins Jahr 2017. Für die Jahre 2018 bis 2020 werden für alle Gemeinden die Werte von 2017 verwendet. Dank der Sortierung nach Gemeinden und Jahren werden die fehlenden Daten mit ffill() nachgereicht.

```python
df = df.ffill()
```

```python
#Stichprobe, ob das geklappt hat
df[(df['gdenr'] == 4077) & (df['jahr'] > 2013)]
```

```python
#Die Spalten Kapital und Einkommen werden nicht mehr benötigt
del df['k']
del df['y']
```

Nun werden die Daten gruppiert (nach Jahr und Kanton) und gewichtet. Dazu dient die Funktion 'gewichter'. Sie nimmt drei Argumente entgegen: Das Dataframe und die Namen der Spalten der Steuerbelastung und des Gewichts. Jede Belastung wird mit dem jeweiligen Gewicht multipliziert. Davon wird die Summe genommen. Diese wird durch die Summe aller Gewichte geteilt und auf zwei Nachkommastellen gerundet.

```python
def gewichter(df, steuern, gewicht):
    s = df[steuern]
    g = df[gewicht]
    
    return ((s * g).sum() / g.sum())

df_final = pd.DataFrame(df.groupby(['jahr', 'kanname']).apply(gewichter, 'etr', 'nettosteuer_fr'))
df_final.reset_index(inplace=True)
df_final.rename(columns={0: 'etr_mean'}, inplace=True)
df_final.set_index('kanname', inplace=True)
```

```python
df_final.head()
```

```python
df_final.sort_values(by='etr_mean', inplace=True)
```

```python
df_final[df_final['jahr'] == 2020][['etr_mean']].plot(kind='bar', figsize=(20, 10))
```

```python
#Export für Q-Tool
df_final[df_final['jahr'] == 2020][['etr_mean']].to_csv('Ergebnisse/etr_juristische_2020.csv')
```
